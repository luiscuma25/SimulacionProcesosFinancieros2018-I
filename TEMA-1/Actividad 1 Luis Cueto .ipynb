Realizar la gráfica del cambio de energía cinética y potencial en este sistema con respecto al tiempo. Recordar que la energía total del sistema esta dada por 
    \begin{equation}
    E(t) = T(t) + V(t) = \frac{1}{2} m [\dot{x}(t)]^2 + \frac{1}{2} k [x(t)]^2
    \end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
masa = 2.5 # kg
t = np.linspace(0, 10) # tiempo de 0 a 10 
k = 3  #constante elástica [N]/[m] 
omega_0 = np.sqrt(k/masa)
x_0 = .5
x_0_dot = .1 
```

```python
x_t_dot = -omega_0 * x_0 * np.sin(omega_0 * t) + x_0_dot * np.cos(omega_0 * t)
##  x normal posición, x punto velocidad, x punto punto aceleración.
```

```python
x_t = x_0 *np.cos(omega_0 *t) + (x_0_dot/omega_0) * np.sin(omega_0 *t)
```

```python
T_t = 0.5 * masa * (x_t_dot)**2 #** es para exponencial.
```

```python
V_t = 0.5 * k * (x_t)**2
```

```python
plt.plot(t, T_t) #tiempo y energía
plt.show()
```

###  - Energía potencial

```python
plt.plot(t, V_t) #tiempo y energía potencial
plt.show()
```

```python
plt.plot(t, T_t, "r-") #tiempo y energía
plt.plot(t, V_t)
plt.show()
```

aqui podemos observar el comportamiento de la energia potencialde una manera mas facil en donde nme base en la formula escrita por latex y defini cada variable con valores para despues sin necesidad de realizar la ecuacion diferencial, ya podemos observar como se conporta en la grafica porqu todo esta definido.
